```python
import requests
from bs4 import BeautifulSoup

API_KEY = "6346700a130067e0e7fb49f26f066733"

# ScraperAPI URL 설정
scraperapi_url = "http://api.scraperapi.com"
params = {
    "api_key": API_KEY,
    "url": "https://kr.investing.com/news/cryptocurrency-news",
    "render": "true"
}

# User-Agent 추가
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

# 요청 보내기
try:
    response = requests.get(scraperapi_url, params=params, headers=headers, timeout=120)
    response.raise_for_status()

    # BeautifulSoup으로 HTML 분석
    try:
        soup = BeautifulSoup(response.content, "lxml")  # 여기에서 lxml 파서를 사용
    except ImportError:
        print("lxml 파서를 찾을 수 없음. 기본 html.parser를 사용합니다.")
        soup = BeautifulSoup(response.content, "html.parser")  # 대체 파서 사용

    print("크롤링 성공! HTML 가져왔습니다.")

    # 뉴스 제목 추출 시도
    news_titles = soup.find_all("a", {"data-test": "article-title-link"})
    if news_titles:
        print("\n추출된 뉴스 제목:")
        for idx, title in enumerate(news_titles, start=1):
            print(f"{idx}. {title.get_text(strip=True)}")
    else:
        print("뉴스를 찾을 수 없습니다.")
except requests.exceptions.RequestException as e:
    print(f"요청 실패: {e}")
```
